# Теоретическая часть

#### Вопрос №1
Что такое NumPy и зачем он используется в анализе данных?

##### Ответ: 
Библиотека для работы с массивами и мат. функциями - c этой же целью используется в анализе данных. Удобен при использовании для обработки больших объемов данных.

#### Вопрос №2
Каковы характеристики и основные типы данных основного объекта массива в NumPy?

##### Ответ: 
Основной объект - ndarray. Типы данных: int, float (с разной "разрядностью", как и int), bool, string. 

#### Вопрос №3
Что такое Matplotlib.PyPlot и для чего он используется при визуализации данных?

##### Ответ: 
Matplotlib.PyPlot - часть библиотеки Matplotplib с помощью которой можно визуализровать данные. Используется для того, чтобы соответственно визуализировать данные, т.к. так данные проще и нагляднее воспринимаются.

#### Вопрос №4
Опишите различные типы графиков, которые можно создать с помощью Matplotlib.PyPlot, и объясните, для чего они нужны.

##### Ответ:
1. Круговая диаграмма (pie) - отображает процентное (количественное) соотношение категорий.
2. Точечный график (scatter) - отображает связь между переменными. Помогает увидеть связь одной переменной от другой.
3. Столбчатая диаграмма (bar) - отображает сравнение величин разных категорий. Например, кол-во заказов за каждый месяц.
4. Линейный график (plot) - отображает изменение значения в зависимости от другой переменной (например времени).

#### Вопрос №5
Что такое описательная статистика и какие статистические характеристики она включает?

##### Ответ:
Набор методов для краткого описания данных. Характеристики:
Среднее (mean) - среднее арифметическое.
Мода (mode) - наиболее часто встречающееся.
Размах (range) - разница между минимальным и максимальным значениями.
Дисперсия (var) - мера разброса данных относительно среднего.

#### Вопрос №6
Как применяются статистические меры в реальных задачах анализа данных? 

##### Ответ:
Примеры. 
1. Выжившие на Титанике - можно вычислить средний возраст (mean) выживших и погибших, чтобы понять, была ли разница. Размах покажет разницу ежду самыми молодыми и самыми старыми выжившими.
2. Сердечно-сосудистые заболевания - можно вычислить среднее давление у пациентов с сердечными заболеваниями и без. С помощью моды определить какое давление встречается чаще всего среди курильщиков.
3. Сайт знакомств - можно посчитать средний возраст людей, получающих больше всего "свайпов".

# Задача №3
Исследовательский центр "Эко-Инсайт" занимается анализом экологической ситуации в различных регионах. Они собрали данные по уровню загрязнения воздуха в крупных городах за последний год и хотят визуализировать эти данные для дальнейшего анализа и представления результатов своим партнёрам и общественности.

#### Формат входных данных
Город: Название города.
Месяц: Месяц измерения уровня загрязнения.
Уровень PM2.5: Уровень частиц PM2.5 в микрограммах на кубический метр воздуха.
Уровень PM10: Уровень частиц PM10 в микрограммах на кубический метр воздуха.
Температура: Среднемесячная температура в градусах Цельсия.

#### Задача
Используя предоставленные данные, создайте следующие графики, которые помогут проанализировать уровень загрязнения воздуха:
1. Создайте линейный график, на котором будет отображено изменение уровней PM2.5 и PM10 по месяцам. Используйте разные цвета для каждого типа частиц.

2. Постройте гистограмму, чтобы показать распределение уровней PM10 по месяцам. Это поможет определить, в каких месяцах чаще всего фиксируются высокие уровни загрязнения.

3. Для выбранного города создайте круговую диаграмму, показывающую долю месяцев с уровнем PM2.5 выше среднего по сравнению с месяцами с уровнем PM2.5 ниже среднего.


#### Dataset
| Месяц | Уровень PM2.5 | Уровень PM10 | Температура |
| --- | --- | --- | --- |
| Январь | 35 | 50 | \-3 |
| Февраль | 30 | 45 | \-1 |
| Март | 25 | 35 | 5 |
| Апрель | 20 | 30 | 10 |
| Май | 15 | 25 | 15 |

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
```

```python
# 1. Линейный график

months, pm_25, pm_10 = df['Месяц'], df['Уровень PM2.5'], df['Уровень PM10']
plt.plot(months, pm_25, color = 'red')
plt.plot(months, pm_10, color = 'blue')
plt.title('Изменение уровней PM2.5 и PM10 по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Уровень загрязнения')
plt.legend(labels=['PM2.5', 'PM10'])
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Гистограмма 

plt.bar(months, pm_10, color = 'orange')
plt.title('Распределение уровней PM10 по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Уровень PM10')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Круговая диаграмма
mean_pm25 = df['Уровень PM2.5'].mean()
above_mean = df[df['Уровень PM2.5'] > mean_pm25].shape[0]
below_mean = df[df['Уровень PM2.5'] < mean_pm25].shape[0]

plt.pie([above_mean, below_mean], labels=['Больше среднего', 'Меньше среднего'])
plt.title('Доля месяцев с уровнем PM2.5 выше и ниже среднего')
plt.show()
```
